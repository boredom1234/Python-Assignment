```python
import pandas as pd
import numpy as np
csv = pd.read_csv('./Parking.csv')
print(csv.info())
print(csv.shape)
print(csv.describe())
```

```python
df_deleted = csv.dropna()
print("DataFrame after deleting rows with missing data:")
print(df_deleted)

df_replaced = csv.fillna(value='replacement_value')
print("DataFrame after replacing missing data:")
print(df_replaced)

df_ffill = csv.ffill()
print("DataFrame after forward filling missing data:")
print(df_ffill)

df_bfill = csv.bfill()
print("DataFrame after backward filling missing data:")
print(df_bfill)
```

```python
grouped = csv.groupby('Registration State')
filtered_group = grouped.get_group('AZ')
print(filtered_group)
```
